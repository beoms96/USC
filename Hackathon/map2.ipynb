```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('ing_7.csv')
tt=df.groupby(['AREA NAME','Hour']).count().reset_index()
i=1
lists=[]
fig=plt.figure(figsize=(24,24))
for name in tt['AREA NAME'].drop_duplicates():
    
    sorted_list=tt[tt['AREA NAME']==name].drop_duplicates()
    fig=plt.subplot(7,3,i)
    rs=pd.DataFrame(sorted_list)
    plt.plot(list(rs['Hour']),list(rs['TIME OCC']),data=list(rs))
    plt.title(name)
    i=i+1

plt.show()
```
